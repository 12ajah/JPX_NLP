```python
import sys
sys.path.append('../src/')

from backtest import Backtest
from IPython.display import display
import pandas as pd

#最大表示列数の指定（ここでは50列を指定）
pd.set_option('display.max_columns', 50)
```

```python
data_dir = "../data/"

# バックテストに必要なデータを取得します。
backtest_codes, backtest_price = Backtest.prepare_data(data_dir)
```

```python
#backtest_price
stock_price = pd.read_csv('../data/new_stock_price_all.csv', index_col=0)
stock_price.set_index('EndOfDayQuote Date', inplace=True)
stock_price.index = pd.to_datetime(stock_price.index)
display(stock_price)
```

```python
backtest_price = pd.concat([backtest_price, stock_price], axis=0)
backtest_price = backtest_price.sort_values(['Local Code', 'EndOfDayQuote Date'])
display(backtest_price.head())
display(backtest_price.tail())
```

```python
portofolio_file_path = "../ipynb/result.csv"
df_submit = Backtest.load_submit(portofolio_file_path)
display(df_submit.head())
```

```python
df_results, df_stocks = Backtest.run(df_submit, backtest_codes, backtest_price)
```

```python
display(df_results.T)
```
